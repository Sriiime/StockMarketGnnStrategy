```python
# Set strategy parameters
short_window = 50
long_window = 200

# Calculate moving averages
df['Short_MA'] = df['Previous Close'].rolling(window=short_window).mean()
df['Long_MA'] = df['Previous Close'].rolling(window=long_window).mean()

# Create signals
df['Signal'] = np.where(df['Short_MA'] > df['Long_MA'], 1, 0)

# Calculate strategy returns
df['Strategy_Returns'] = df['Previous Close'].pct_change() * df['Signal'].shift(1)

# Evaluate strategy performance
sharpe_ratio = df['Strategy_Returns'].mean() / df['Strategy_Returns'].std() * np.sqrt(252)
sortino_ratio = df['Strategy_Returns'].mean() / df['Strategy_Returns'][df['Strategy_Returns'] < 0].std() * np.sqrt(252)
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Sortino Ratio: {sortino_ratio:.2f}")
```

```python
# Plot strategy performance
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Previous Close'], label='Close Price')
plt.plot(df['Previous Close'] * df['Signal'], label='Strategy')
plt.legend()
plt.show()

# Refine strategy
def optimize_strategy_params(df, short_window_range, long_window_range):
    def objective(params):
        short_window, long_window = np.int_(params)
        df['Short_MA'] = df['Previous Close'].rolling(window=short_window).mean()
        df['Long_MA'] = df['Previous Close'].rolling(window=long_window).mean()
        df['Signal'] = np.where(df['Short_MA'] > df['Long_MA'], 1, 0)
        df['Strategy_Returns'] = df['Previous Close'].pct_change() * df['Signal'].shift(1)
        return -df['Strategy_Returns'].mean() / df['Strategy_Returns'].std()

    bounds = [(int(short_window_range[0]), int(short_window_range[1])),
              (int(long_window_range[0]), int(long_window_range[1]))]
    res = minimize(objective, [short_window, long_window], bounds=bounds)
    return res.x

optimal_params = optimize_strategy_params(df, (20, 100), (100, 300))
print(f"Optimal Parameters: {optimal_params}")
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
```

```python
# Create industry graph
G = nx.Graph()
node_map = {}  # Map node symbols to indices
for i, (index, row) in enumerate(df.iterrows()):
    G.add_node(row['Symbol'], industry=row['Industry'])
    node_map[row['Symbol']] = i

industry_map = {industry: i for i, industry in enumerate(df['Industry'].unique())}
num_nodes = len(G.nodes)
num_features = 16  # Adjusted feature dimension


node_features = torch.randn(num_nodes, num_features, dtype=torch.float)  # Random features


edge_list = []
for node1 in G.nodes:
    for node2 in G.nodes:
        if node1 != node2:
            edge_list.append((node_map[node1], node_map[node2]))  # Use node indices

edge_index = torch.tensor(edge_list, dtype=torch.long).t().contiguous()


# Create PyTorch Geometric data object
data = Data(x=node_features, edge_index=edge_index)


# Define GCN model
class GCN(nn.Module):
    def __init__(self):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_features, 32)
        self.conv2 = GCNConv(32, len(industry_map))

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = torch.relu(self.conv1(x, edge_index))
        x = self.conv2(x, edge_index)
        return x


# Initialize model, optimizer, and loss function
model = GCN()
optimizer = optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.CrossEntropyLoss()


# Train model
for epoch in range(100):
    optimizer.zero_grad()
    out = model(data)
    loss = loss_fn(out, torch.tensor([industry_map[G.nodes[node]['industry']] for node in G.nodes], dtype=torch.long))
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item():.4f}')


# Evaluate model
model.eval()
_, predicted = torch.max(model(data), dim=1)
print('Industry Clustering Results:')
print(predicted)


# Visualize industry clusters
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
colors = ['blue' if industry == predicted[i].item() else 'red' for i, industry in enumerate(node_features)]
nx.draw_networkx(G, pos, node_color=colors)
plt.show()

```

```python
# Calculate correlations
corr_data = df[['Open', 'High', 'Low', 'Previous Close', 'Last Traded Price']].corr()
```

```python
# Create weighted graph
G = nx.Graph()
for i in range(len(corr_data.columns)):
    for j in range(i+1, len(corr_data.columns)):
        G.add_edge(corr_data.columns[i], corr_data.columns[j], weight=corr_data.iloc[i, j])

# Calculate centrality measures
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
pagerank_centrality = nx.pagerank(G)

# Combine centrality measures
centrality_measures = pd.DataFrame({
    'Stock': list(G.nodes),
    'Degree Centrality': [degree_centrality[node] for node in G.nodes],
    'Closeness Centrality': [closeness_centrality[node] for node in G.nodes],
    'Betweenness Centrality': [betweenness_centrality[node] for node in G.nodes],
    'Eigenvector Centrality': [eigenvector_centrality[node] for node in G.nodes],
    'PageRank Centrality': [pagerank_centrality[node] for node in G.nodes]
})

# Sort by centrality measures
centrality_measures = centrality_measures.sort_values(by='Eigenvector Centrality', ascending=False)

# Visualize top 10 influential stocks
top_10_stocks = centrality_measures.head(10)
print(top_10_stocks)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(centrality_measures)
```

```python
# Convert numeric columns to float
numeric_cols = ['52 Week High', '52 Week Low', '30 Day Percentage Change', '365 Day Percentage Change']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Remove rows with missing values
df = df.dropna(subset=['52 Week High', '52 Week Low', 'Industry', 'Company Name'])

# Calculate indicators
df['SMA_50'] = df['52 Week High'].rolling(window=50).mean()
df['SMA_200'] = df['52 Week High'].rolling(window=200).mean()
df['RSI'] = 100 - (100 / (1 + df['30 Day Percentage Change']))
df['MACD'] = df['52 Week High'].ewm(span=12, adjust=False).mean() - df['52 Week High'].ewm(span=26, adjust=False).mean()

# Define signal conditions
def generate_signal(row):
    if (row['52 Week High'] > row['SMA_200'] and
        row['RSI'] < 70 and
        row['MACD'] > 0 and
        row['30 Day Percentage Change'] > 0):
        return 'Buy'
    elif (row['52 Week Low'] < row['SMA_50'] and
          row['RSI'] > 30 and
          row['MACD'] < 0 and
          row['30 Day Percentage Change'] < 0):
        return 'Sell'
    else:
        return 'Hold'

df['Signal'] = df.apply(generate_signal, axis=1)

# Plot signals by industry
plt.figure(figsize=(12, 8))
sns.countplot(x='Industry', hue='Signal', data=df)
plt.xlabel('Industry')
plt.ylabel('Count')
plt.title('Signals by Industry')
plt.xticks(rotation=45)
plt.legend(title='Signal')
plt.tight_layout()
plt.show()

```
